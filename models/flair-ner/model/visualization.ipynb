```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the TSV file
df = pd.read_csv("loss.tsv", delimiter="\t")

df.head()
```

```python
# Extract EPOCHS and other columns
epochs = df["EPOCH"]
bad_epochs = df["BAD_EPOCHS"]
learning_rate = df["LEARNING_RATE"]
train_loss = df["TRAIN_LOSS"]
dev_loss = df["DEV_LOSS"]
dev_precision = df["DEV_PRECISION"]
dev_recall = df["DEV_RECALL"]
dev_f1 = df["DEV_F1"]
dev_accuracy = df["DEV_ACCURACY"]
```

```python
plt.plot(epochs, train_loss, label="Train Loss")
plt.plot(epochs, dev_loss, label="Development Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Flair NER Loss Curves")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(epochs, dev_accuracy, label="Development Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Flair NER Accuracy Curve")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(epochs, dev_precision, label="Development Precision")
plt.plot(epochs, dev_recall, label="Development Recall")
plt.plot(epochs, dev_f1, label="Development F1-score")
plt.xlabel("Epochs")
plt.ylabel("Metrics")
plt.title("Flair NER Precision, Recall, and F1-Score Curves")
plt.legend()
plt.grid(True)
plt.show()
```
